```python
# Open the lammps rdf file
with open('system.rdf', 'r') as f:
    data = f.readlines()
```

```python
# Read all the lines and get the x and y values
x = []
y = []
for l in data[200:]:
    s = l.split()
    x.append(float(s[1]))
    y.append(float(s[2]))
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```
